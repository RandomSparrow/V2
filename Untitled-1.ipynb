```python
import pydicom
from pydicom.pixel_data_handlers.util import apply_voi_lut
from PIL import Image
import numpy as np
import os

#Funkcja zamiany dcm na jpg
def dicom_to_jpg(dicom_file, jpg_file):
    
    ds = pydicom.dcmread(dicom_file)
    data = apply_voi_lut(ds.pixel_array, ds)

    if ds.PhotometricInterpretation == "MONOCHROME1":
        data = np.amax(data) - data

    data = (data - np.min(data)) / (np.max(data) - np.min(data)) * 255
    data = data.astype(np.uint8)

    image = Image.fromarray(data)
    image.save(jpg_file)
```

```python
import os
from glob import glob

#Zamiana plików dcm na jpg
dcm_path = "C:\\Users\\Marcel\\Desktop\\Studia\\Przetwarzanie obrazów\\Projekt\\V2\\Data_dcm/*"
jpg_path = "C:\\Users\\Marcel\\Desktop\\Studia\\Przetwarzanie obrazów\\Projekt\\V2\\Data_jpg"
image_list = glob(dcm_path)

for image in image_list:
    image_name = os.path.basename(os.path.normpath(image))
    dicom_to_jpg(image, os.path.join(jpg_path, image_name + '.jpg'))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage as ski

#Funkcja segmentacji
def segmentate(image_path, mask_size, save_path):
    image = ski.io.imread(image_path)

    blurred = cv2.GaussianBlur(image, (5, 5), 0)

    _, binary = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    kernel = np.ones((3, 3), np.uint8)
    closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel, iterations=2)
    opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel, iterations=4)

    binary_image_inverted = cv2.bitwise_not(opening)

    segmented = binary + binary_image_inverted

    _, segmented = cv2.threshold(segmented, 0, 1, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    height = 256
    width = 256

    x, y = np.ogrid[:height, :width]
    mask = (x - height / 2)**2 + (y - width / 2)**2 < (mask_size)**2
    mask = mask.astype(int)

    segmented[mask==0]=1

    kernel2 = np.ones((2, 2), np.uint8)
    closing2 = cv2.morphologyEx(segmented, cv2.MORPH_CLOSE, kernel2, iterations=1)

    segmented_inverted = cv2.bitwise_not(closing2)

    fig, ax = plt.subplots(figsize=(2.78, 2.78))
    ax.imshow(segmented_inverted, cmap='gray')
    plt.axis(False)
    plt.savefig(save_path, dpi = 120, bbox_inches='tight', pad_inches=0)
```

```python
#Zmiana ilości wymiarów na 2D
import numpy as np
from PIL import Image

def rgb_to_grayscale(rgb_img):
    grayscale_img = np.dot(rgb_img[..., :3], [0.2126, 0.7152, 0.0722])
    return grayscale_img


jpg_file_folder = "C:\\Users\\Marcel\\Desktop\\Studia\\Przetwarzanie obrazów\\Projekt\\V2\\Segmentated/*"
file_folder = "C:\\Users\\Marcel\\Desktop\\Studia\\Przetwarzanie obrazów\\Projekt\\V2\\Segmentatedx2"
image_list = glob(jpg_file_folder)

for file in image_list:
    image_name = os.path.basename(os.path.normpath(file))
    image = ski.io.imread(file)
    rgb_image = np.array(image)
    grayscale_image = rgb_to_grayscale(rgb_image)
    Image.fromarray(grayscale_image.astype(np.uint8)).save(os.path.join(file_folder, image_name + '.jpg'))
```
